# 01. 인형뽑기문제

```python
from pandas import DataFrame
def solution(board,moves):
    answer,basket = 0,[]
    df=DataFrame(board)
    for c in moves:
        for i in range(len(df)):
            if df[c-1][i] != 0:
                basket.append(df.iloc[i][c-1])
                df[c-1][i] = 0
                break
    for w in range(len(basket)):
        for i in range(len(basket)):
            if i<len(basket)-1:
                if basket[i] == basket[i+1]:
                    del basket[i]
                    del basket[i]
                    answer += 1
    return answer*2
```
