# 문제 2

```python
def solution2(arr):
    cnt = 0
    cnt_list = []
    for i in range(len(arr)):
        for j in range(len(arr)):
            if (arr[i] == arr[j]):
                cnt+=1
        if(arr[i]==cnt):
            cnt_list.append(arr[i])
        cnt = 0
    cnt_list.sort()
    if(len(cnt_list)!=0):
        answer = cnt_list[-1]
    else:
        answer = 0
    return answer
```
